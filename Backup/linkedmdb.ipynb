```python
import requests

def query_wikidata_rdf(entity_id):
    # Wikidata API endpoint for RDF
    api_url = f"https://www.wikidata.org/wiki/Special:EntityData/{entity_id}.rdf"

    # Make the request to Wikidata API for RDF
    response = requests.get(api_url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Return the RDF data
        return response.text
    else:
        # If the request was not successful, print the error code
        print(f"Error: {response.status_code}")
        return None

# Example query for the movie "Inception" (entity ID: Q163872)
movie_entity_id = "Q163872"
rdf_data = query_wikidata_rdf(movie_entity_id)

# Display the RDF data
print(rdf_data)
```
